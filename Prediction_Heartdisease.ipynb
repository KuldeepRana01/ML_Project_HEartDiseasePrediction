**# PREDICTION BY LINEAR REGRESSION MODEL**

```python
# @title Default title text
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# Load Dataset
df = pd.read_csv("/content/drive/MyDrive/heart_disease_data.csv")

# Define Features (X) and Target (Y)
X = df.drop(columns=["target"])  # Features (independent variables)
Y = df["target"]  # Target (1 = Heart Disease, 0 = No Heart Disease)

# Split Data into Train & Test Sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardize the Data (Improves Linear Regression Performance)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, Y_train)

# Function to Predict Heart Disease
def predict_heart_disease(input_data):
    """Predicts heart disease based on user input."""
    arr = np.asarray(input_data).reshape(1, -1)  # Convert input to array
    arr = scaler.transform(arr)  # Scale input data
    prediction = model.predict(arr)

    return "The person has heart disease." if prediction[0] > 0.5 else "The person does not have heart disease."

# Function to take validated input
def get_valid_input(prompt, min_value, max_value, datatype=int):
    while True:
        try:
            value = datatype(input(prompt))
            if value < min_value or value > max_value:
                print(f"Please enter a value between {min_value} and {max_value}.")
            else:
                return value
        except ValueError:
            print(f"Invalid input. Please enter a {datatype.__name__}.")

# Take User Input with Validation
age = get_valid_input("Enter age: ", 1, 120)  # Assuming age between 1-120
sex = get_valid_input("Enter sex (0 for female, 1 for male): ", 0, 1)
cp = get_valid_input("Enter chest pain type Value (0-3) (0: typical angina, 1: atypical angina, 2: non-anginal pain, 3: asymptomatic): ", 0, 3)
trestbps = get_valid_input("Enter resting blood pressure (Range value: 29-77): ", 29, 77)
chol = get_valid_input("Enter serum cholesterol in mg/dl (Range value: 126-564): ", 126, 564)
fbs = get_valid_input("Enter fasting blood sugar > 120 mg/dl (1 = true; 0 = false): ", 0, 1)
restecg = get_valid_input("Enter resting electrocardiographic results (0-2): ", 0, 2)
thalach = get_valid_input("Enter maximum heart rate achieved (Range: 71-202): ", 71, 202)
exang = get_valid_input("Enter exercise-induced angina (1 = yes; 0 = no): ", 0, 1)
oldpeak = get_valid_input("Enter ST depression induced by exercise (Range value: 0.0 - 6.2): ", 0.0, 6.2, datatype=float)
slope = get_valid_input("Enter the slope of the peak exercise ST segment (0 = upsloping, 1 = flat, 2 = downsloping): ", 0, 2)
ca = get_valid_input("Enter number of major vessels colored by fluoroscopy (0-4): ", 0, 4)
thal = get_valid_input("Enter thalassemia type (1 = normal, 2 = fixed defect, 3 = reversible defect): ", 1, 3)

# Make Prediction
input_data = (age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal)
result = predict_heart_disease(input_data)
print("\nPrediction:", result)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

# Load Dataset
df = pd.read_csv("/content/drive/MyDrive/heart_disease_data.csv")  # Changed to the full file path

# Define Features (X) and Target (Y)
X = df.drop(columns=["target"])  # Features (independent variables)
Y = df["target"]  # Target (1 = Heart Disease, 0 = No Heart Disease)

# ... (rest of the code remains the same) ...

# Split Data into Train & Test Sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Standardize the Data (Improves Decision Tree Performance)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train Decision Tree Model
model = DecisionTreeClassifier(max_depth=5, random_state=2)
model.fit(X_train, Y_train)

# Function to Predict Heart Disease
def predict_heart_disease(input_data):
    """Predicts heart disease based on user input using Decision Tree."""
    arr = np.asarray(input_data).reshape(1, -1)  # Convert input to array
    arr = scaler.transform(arr)  # Scale input data
    prediction = model.predict(arr)

    return "The person has heart disease." if prediction[0] == 1 else "The person does not have heart disease."

# Take User Input
age = int(input("Enter age: "))
sex = int(input("Enter sex (0 for female, 1 for male): "))
cp = int(input("Enter chest pain type Value(1: Typical angina/2: Atypical angina/3: Non-anginal pain)(0-3): "))
trestbps = int(input("Enter resting blood pressure(Range value:29-77): "))
chol = int(input("Enter serum cholesterol in mg/dl(Range value:126-564): "))
fbs = int(input("Enter fasting blood sugar > 120 mg/dl (1 = true; 0 = false): "))
restecg = int(input("Enter resting electrocardiographic results(0 = Normal/ 1 = ST-T wave abnormality/ 2 = Probable left ventricular hypertrophy)(0-2): "))
thalach = int(input("Enter maximum heart rate achieved(71-202): "))
exang = int(input("Enter exercise-induced angina (1 = yes; 0 = no): "))
oldpeak = float(input("Enter ST depression induced by exercise(Range value: 0.0 - 6.2): "))
slope = int(input("Enter the slope of the peak exercise ST segment(0 =Upsloping/ 1 =Flat/ 2 =Downsloping)(0-2): "))
ca = int(input("Enter number of major vessels (0-3) colored by fluoroscopy(0-4): "))
thal = int(input("Enter thalassemia type(1=normal, 2=fixed defect, 3=reversible defect): "))

# Make Prediction
input_data = (age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal)
result = predict_heart_disease(input_data)
print("\nPrediction:", result)
```
